Crea una serie de 10 elementos, que contenga números enteros aleatorios del 70 al 100, que representan las notas mensuales
de los examenes de esta asignatura para un alumno. Configura el índice para que sean los nombres de los meses, comenzando en septiembre y terminando 
en junio. 

En base a esta serie, responde las siguientes preguntas:
- ¿Cuál es la nota media del estudiante en las pruebas durante todo el año?
- ¿Cuál es la nota media del estudiante en las pruebas durante la primera mitad del año (es decir, los primeros cinco meses)?
- ¿Cuál es la nota media del estudiante en las pruebas durante la segunda mitad del año?
- ¿El estudiante mejoró sus notas en la segunda mitad? Si es así, ¿en cuánto?

```python
import numpy as np
import pandas as pd
serieNotas = pd.Series(np.random.randint(70,101,size = 10),index="septiembre octubre noviembre diciembre enero febrero marzo abril mayo junio".split(" "))
 
print(serieNotas)
mediaAnio = serieNotas.mean()
mediaAnio1 = serieNotas.head(5).mean()
mediaAnio2 = serieNotas.tail(5).mean()
 
print(f"Media año {mediaAnio}")
print(f"Media primera parte año {mediaAnio1}")
print(f"Media segunda parte año {mediaAnio2}")
if mediaAnio2 > mediaAnio1:
    print(f"El alumno mejoro {mediaAnio2 - mediaAnio1} puntos")
```

Cuando estaba en el instituto y en la universidad, nuestros profesores a veces daban exámenes que eran extremadamente difíciles. En lugar de suspender 
a la mayor parte de la clase, a veces "escalaban" las notas de los exámenes. Es decir, asumían que la puntuación media de la prueba debería ser 85. 
Por lo tanto, calculaban la diferencia entre nuestra media real y esta media ideal, y luego sumaban puntos a cada una de nuestras notas.

Para este ejercicio, quiero que generes 10 puntuaciones de exámenes entre 40 y 60 usando un índice, comenzando en septiembre y terminando en junio. 
Encuentra la media de las puntuaciones y suma la diferencia entre la media y 85 para cada una de las puntuaciones.

```python
serieNotas = pd.Series(np.random.randint(40,61,size = 10),index="septiembre octubre noviembre diciembre enero febrero marzo abril mayo junio".split(" "))

media = serieNotas.mean()
print(serieNotas + (85 - media))
```
