Concatena estos DataFrames verticalmente.

````python
import pandas as pd

df1 = pd.DataFrame({
    'ID': ['A1', 'A2', 'A3'],
    'Sales': [100, 200, 300]
})

df2 = pd.DataFrame({
    'ID': ['A4', 'A5', 'A6'],
    'Revenue': [400, 500, 600]
})

print(pd.concat([df1,df2],axis=0))
````

Usa los siguientes DataFrames y realiza una concatenación utilizando join='inner' y otra utilizando join='outer'. Compara los resultados.

```python
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

df2 = pd.DataFrame({
    'A': [10, 11, 12],
    'D': [13, 14, 15]
})
```

````python
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

df2 = pd.DataFrame({
    'A': [10, 11, 12],
    'D': [13, 14, 15]
})

print("INNER")
print(pd.concat([df1,df2],join="inner"))
print("OUTER")
print(pd.concat([df1,df2],join="outer"))
````

Usa los siguientes DataFrames y realiza un merge utilizando `left_on` y `right_on`. Luego realiza un `outer_join` y muestra el resultado.

```python
df1 = pd.DataFrame({
    'Product_ID': ['P1', 'P2', 'P3'],
    'Sales': [100, 200, 300]
})

df2 = pd.DataFrame({
    'ID': ['P1', 'P3', 'P4'],
    'Discount': [10, 15, 20]
})
```

````python
df1 = pd.DataFrame({
    'Product_ID': ['P1', 'P2', 'P3'],
    'Sales': [100, 200, 300]
})

df2 = pd.DataFrame({
    'ID': ['P1', 'P3', 'P4'],
    'Discount': [10, 15, 20]
})

print(pd.merge(df1,df2,how='outer',left_on="Product_ID",right_on="ID"))
````

Usa los siguientes DataFrames y realiza un merge utilizando múltiples claves:

```python
df1 = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Region': ['North', 'South', 'East'],
    'Sales': [250, 150, 300]
})

df2 = pd.DataFrame({
    'Product': ['A', 'B', 'D'],
    'Region': ['North', 'East', 'West'],
    'Profit': [50, 60, 70]
})
```

Realiza un inner join utilizando las claves `Product` y `Region`.

````python
df1 = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Region': ['North', 'South', 'East'],
    'Sales': [250, 150, 300]
})

df2 = pd.DataFrame({
    'Product': ['A', 'B', 'D'],
    'Region': ['North', 'East', 'West'],
    'Profit': [50, 60, 70]
})

print(pd.merge(df1,df2,how='inner',on=["Product","Region"]))
````

Concatena estos DataFrames verticalmente utilizando `join='outer'`. Luego, reemplaza todos los valores nulos con la media de la columna correspondiente.

````python
df1 = pd.DataFrame({
    'A': [1, 2, None],
    'B': [4, None, 6]
})

df2 = pd.DataFrame({
    'A': [None, 8, 9],
    'C': [10, 11, 12]
})

concate = pd.concat([df1,df2],join='outer',axis=0,ignore_index=True)
media = concate.mean()
concate = concate.fillna(media)
print(concate)
print(media)
````
