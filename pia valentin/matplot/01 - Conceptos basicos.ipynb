Crea un gráfico de líneas que muestre la temperatura media mensual de una ciudad a lo largo de un año. Personaliza el gráfico con colores, marcadores y etiquetas.

```python
import matplotlib.pyplot as plot
import numpy as np

meses = ["enero","febrero","marzo","abril","mayo","junio","julio","agosto","septiembre","octubre","noviembre","diciembre"]
tMes = np.random.randint(10,30,size=12) + np.random.random(size=12)

plot.plot(meses,tMes)
plot.xticks(rotation=25)
plot.show()
```

Combina un gráfico de líneas y un gráfico de puntos para visualizar la evolución de las ventas mensuales de un producto a lo largo del año mediante un gráfico de líneas, y resaltar los datos individuales de cada mes utilizando puntos. Crea un gráfico de líneas que muestre la evolución de las ventas trimestrales de una empresa durante el año. Usa los siguientes datos:

- Trimestres: 'Q1', 'Q2', 'Q3', 'Q4'
- Ventas (en miles de unidades): 50, 80, 70, 90

Personaliza el gráfico con:

- Un título que diga "Ventas Trimestrales de la Empresa".
- Etiquetas en los ejes X y Y.
- Un color y estilo de línea personalizado.
- Una leyenda que indique que la línea representa las ventas.

```python
trimestres = ['Q1', 'Q2', 'Q3', 'Q4']
ventas = [50, 80, 70, 90]

plot.plot(trimestres,ventas,color="r",linestyle="dashed",label="Ventas")
plot.title("Ventas Trimestrales de la Empresa")
plot.xlabel("Trimestres")
plot.ylabel("Ventas")
plot.legend()
```

Crea un gráfico de puntos que muestre la relación entre el número de horas estudiadas y las calificaciones obtenidas en un examen. Usa los siguientes datos:

- Horas de Estudio: [1, 2, 3, 4, 5]
- Calificaciones: [60, 65, 70, 75, 80]

Personaliza el gráfico con:

- Un título que diga "Relación entre Horas de Estudio y Calificaciones".
- Etiquetas en los ejes X y Y.
- Un color y forma de marcador personalizados para los puntos.
- Una leyenda que indique que los puntos representan las calificaciones.

```python
horasdeEstudio = [1, 2, 3, 4, 5]
calificaciones =  [60, 65, 70, 75, 80]

plot.plot(horasdeEstudio,calificaciones,color='g',marker="h",label="Calificaciones")
plot.title("Relación entre Horas de Estudio y Calificaciones")
plot.xlabel("Horas estudio")
plot.ylabel("Calificaciones")
```

Crea un gráfico de líneas que muestre las temperaturas diarias durante una semana. Usa los siguientes datos:

- Días: 'Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'
- Temperaturas (en grados Celsius): 22, 24, 19, 25, 23, 26, 24

Personaliza el gráfico con:

- Un título que diga "Temperaturas Diarias de la Semana".
- Etiquetas en los ejes X y Y.
- Ajuste de los límites del eje Y para enfocarse entre 15 y 30 grados.
- Etiquetas rotadas para los días de la semana.

```python
dias = 'Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'
temp = 22, 24, 19, 25, 23, 26, 24

plot.plot(dias,temp)
plot.xlabel("Dias")
plot.ylabel("Temperatura")
plot.ylim(15,30)
plot.xticks(rotation=45)
plot.title("Temperaturas Diarias de la Semana")
```
