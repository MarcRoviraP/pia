Crea un gráfico de barras que muestre las calificaciones de los estudiantes en diferentes asignaturas. Personaliza el gráfico cambiando el color de las barras y añadiendo etiquetas en cada barra.

```python
import matplotlib.pyplot as plot
import numpy as np
estudiantes = [f"Alumno {i}" for i in range(1,11)]
notas = [5,9,6,7,2,3,1,5,8,6]

plot.bar(estudiantes,notas,color="y")
plot.xticks(rotation=45)
[plot.text(i,notas[i],notas[i],ha="center") for i in range(len(notas))]
```

Crea un gráfico de barras agrupadas para comparar las ventas mensuales de dos productos diferentes a lo largo de seis meses. Añade etiquetas y leyendas para diferenciar las dos series de datos.

```python
ventasMensualesPlay = np.random.randint(100,300,size=6)
ventasMensualesXBox = np.random.randint(100,300,size=6)
meses = ["Enero","Febrero","Marzo","Abril","Mayo","Junio"]

x = np.arange(len(meses))
width = 0.4
plot.bar(x - width / 2 ,ventasMensualesPlay,width,label="PlaySation",color="b")
plot.bar(x + width / 2 ,ventasMensualesXBox,width,label="XBox",color="g")
plot.title("Playstation vs XBox")
plot.legend()
plot.xticks(x,meses)
plot.ylim(0,400)

[plot.text(i,ventasMensualesPlay[i] + 5,ventasMensualesPlay[i],ha="right") for i in range(len(ventasMensualesPlay))]
[plot.text(i,ventasMensualesXBox[i] + 5,ventasMensualesXBox[i],ha="left") for i in range(len(ventasMensualesXBox))]

plot.ylabel("Ventas")
```

Crea un histograma que muestre la distribución de puntuaciones en un examen. Ajusta el número de intervalos para que se adapten a la variabilidad de las puntuaciones y añade etiquetas en los ejes y un título adecuado.

```python
notas = np.random.randint(11,size=500)
plot.hist(notas,edgecolor="black",bins=5)
plot.ylabel("Cantidad notas repetidas")
plot.xlabel("Notas")
plot.title("Notas repetidas")
```

Genera un histograma con datos de alturas de personas en dos grupos diferentes (por ejemplo, hombres y mujeres) y compáralos en el mismo gráfico. Utiliza diferentes colores y transparencias para distinguir los dos grupos.

```python
alturasHombre = np.random.randint(50,100,size=1000) / 100 + 1
alturasMujer = np.random.randint(50,90,size=1000) / 100 + 1
bins = np.linspace(1.5, 2.0, 20)
plot.hist(alturasHombre,bins=bins,alpha=0.7,edgecolor="black")
plot.hist(alturasMujer,bins=bins,alpha=0.7,edgecolor="black")
```
