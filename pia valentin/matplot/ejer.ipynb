## 2. Un profesor registra las siguientes notas:
[2, 3, 4, 5, 6, 7, 8, 15]

### a) Calcula los cuartiles y la mediana  

Q1 = [2,3,4,5] -> (3 + 4) / 2 = 3.5
Q3 = [6,7,8,15] -> (7 + 8) / 2 = 7.5

Mediana = (5 + 6) / 2 = 5.5

### b) Calcula el IQR  

IQR = Q3 - Q1

7.5 - 3.5 = 4

### c) Detecta si hay outliers  

7.5 + 4 * 1.5 = 13.5

3.5 - 4 * 1.5 = -2.5

Si el 15 esta fuera del rango válido de -2.5 - 13.5 por lo tanto es outliner
