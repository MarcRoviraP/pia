Crea una figura con cuatro subplots (2x2) que muestren diferentes relaciones entre dos conjuntos de datos. Personaliza cada subplot con diferentes colores y estilos de línea.

```python
import matplotlib.pyplot as plot 

x = range(1,11)
y1 = range(1,11)
y2 = range(1,11)[::-1]
y3 = range(1,11)[::-1]
y4 = range(1,11)

plot.figure(figsize=(10,5))

plot.subplot(2,2,1)
plot.plot(x,y1,marker="o")
plot.title("Plot 1")


plot.subplot(2,2,2)
plot.bar(x,y2,color="r")
plot.title("Plot 2")

plot.subplot(2,2,3)
plot.barh(x,y3)
plot.title("Plot 3")

plot.subplot(2,2,4)
plot.scatter(x,y4,color="g")
plot.title("Plot 4")

plot.tight_layout()
```

Genera una figura con tres subplots dispuestos en una columna. El primer subplot debe tener un gráfico de dispersión, el segundo un gráfico de líneas, y el tercero un histograma. Asegúrate de etiquetar claramente los ejes y los datos.

```python
import numpy as np

histList = np.random.randint(100,size=500)

plot.subplot(3,1,1)
plot.scatter(x,y1)
plot.subplot(3,1,2)
plot.plot(x,y2)
plot.subplot(3,1,3)
plot.hist(histList,bins=10,edgecolor="black")
```

Utiliza `GridSpec` para crear una figura con subplots de diferentes tamaños. Por ejemplo, un subplot grande en la parte superior y dos subplots más pequeños en la parte inferior.

```python
import matplotlib.gridspec as gridspec

figura = plot.figure(figsize=(10,10))
gs = gridspec.GridSpec(2,2)
p1 = figura.add_subplot(gs[0,:])
p2 = figura.add_subplot(gs[1,0])
p3 = figura.add_subplot(gs[1,1])

p1.plot(x,y1)
p2.plot(x,y1)
p3.plot(x,y1)
```
