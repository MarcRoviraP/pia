Crea una figura con cuatro subplots (2x2) que muestren diferentes relaciones entre dos conjuntos de datos. Personaliza cada subplot con diferentes colores y estilos de línea.

```python
import matplotlib.pyplot as plot 

x = range(1,11)
y1 = range(1,11)
y2 = range(1,11)[::-1]
y3 = range(1,11)[::-1]
y4 = range(1,11)

plot.figure(figsize=(10,5))

plot.subplot(2,2,1)
plot.plot(x,y1,marker="o")
plot.title("Plot 1")


plot.subplot(2,2,2)
plot.bar(x,y2,color="r")
plot.title("Plot 2")

plot.subplot(2,2,3)
plot.barh(x,y3)
plot.title("Plot 3")

plot.subplot(2,2,4)
plot.scatter(x,y4,color="g")
plot.title("Plot 4")

plot.tight_layout()
```

Utiliza datos de temperaturas medias mensuales para crear un gráfico de área que compare las temperaturas de dos ciudades diferentes a lo largo de un año. Asegúrate de etiquetar claramente los ejes y añadir una leyenda.

```python
valencia = [12,16,18,25,26,27,30,31,35,25,20,18]
madrid = [19,14,20,24,28,30,32,33,38,27,19,16]
x = range(12)

plot.fill_between(x,valencia,label="Valencia",alpha=0.6)
plot.fill_between(x,madrid,label="Madrid",alpha=0.6)

plot.title("Temperaturas Valencia Madrid")
plot.xlabel("Meses")
plot.ylabel("Grados")
plot.legend()
```

Genera un conjunto de datos con cuatro categorías diferentes y crea un gráfico de violín para comparar sus distribuciones. Personaliza los colores y las etiquetas del gráfico.

```python
cat1 = np.random.randint(100,size=100)
cat2 = np.random.randint(100,size=100)
cat3 = np.random.randint(100,size=100)
cat4 = np.random.randint(100,size=100)

datos = [cat1,cat2,cat3,cat4]
violines = plot.violinplot(datos)
colores = ["r", "g", "b", "y"]

for i, violin in enumerate(violines['bodies']):
    violin.set_facecolor(colores[i])
    violin.set_alpha(0.7)
plot.title("Violin")
plot.xlabel("Categorías")
plot.ylabel("Valores")
```

Crea un gráfico de violín con datos simulados que incluya tanto la media como la mediana. Asegúrate de diferenciar visualmente ambos estadísticos en el gráfico.

```python
cat1 = np.random.randint(100,size=100)
cat2 = np.random.randint(100,size=100)
cat3 = np.random.randint(100,size=100)
cat4 = np.random.randint(100,size=100)

datos = [cat1,cat2,cat3,cat4]
violines = plot.violinplot(datos,showmeans=True,showmedians=True)

violines['cmeans'].set_color('blue')
violines['cmedians'].set_color('red')
plot.title("Violin")
plot.xlabel("Categorías")
plot.ylabel("Valores")
```

Utiliza un conjunto de datos real (por ejemplo, de calificaciones de estudiantes o medidas físicas) y visualiza la distribución con un gráfico de violín. Personaliza el gráfico para que sea claro y estéticamente agradable.

```python
from sklearn.datasets import load_iris

# Cargar dataset real
iris = load_iris()
data = iris.data
target = iris.target
labels = iris.target_names

petal_length = data[:, 2]

# Separar por especie
setosa = petal_length[target == 0]
versicolor = petal_length[target == 1]
virginica = petal_length[target == 2]

datos = [setosa,versicolor,virginica]

violines = plot.violinplot(datos, showmeans=True, showmedians=True)

colores = ["lightblue", "lightgreen", "lightcoral"]

for i, violin in enumerate(violines['bodies']):
    violin.set_facecolor(colores[i])
    violin.set_edgecolor('black')
    violin.set_alpha(0.7)

violines['cmeans'].set_color('blue')
violines['cmedians'].set_color('darkred')

# Etiquetas
plot.xticks([1, 2, 3], labels)
plot.title("Distribución de la longitud del pétalo por especie (Iris)")
plot.xlabel("Especie")
plot.ylabel("Longitud del pétalo (cm)")
```
