1. Crea una serie de pandas utilizando un diccionario con las siguientes etiquetas y valores:

    ```python
    diccionario = {"a": 10, "b": 20, "c": 30, "d": 40, "e": 50}
    ```

    Asigna esta serie a la variable serie.

````python
import pandas as pd
diccionario = {"a": 10, "b": 20, "c": 30, "d": 40, "e": 50}
pandasDic = pd.Series(diccionario)

print(pandasDic)
````

2. Utilizando métodos de acceso por etiqueta y posición, extrae y muestra el valor de las siguientes posiciones y etiquetas:

   - Posición 2
   - Etiqueta "d"
   - Etiquetas "b" y "e"

````python
print(pandasDic.iloc[2])
print(pandasDic["d"])
print(pandasDic.loc[["b","e"]])
````

3. Utiliza el método `iloc` para seleccionar los primeros 3 elementos de la serie y el método `loc` para seleccionar los elementos desde "b" hasta "d"

````python
print(pandasDic.iloc[:3])
print(pandasDic.loc["b":"d"])
````

1. Dada la serie creada en el Ejercicio 1, asigna un nuevo índice a la serie usando el siguiente conjunto de etiquetas: ['A', 'B', 'C', 'D', 'E'].


````python
indices = ['A', 'B', 'C', 'D', 'E']
pandasDic.index = indices
print(pandasDic)
````

2. Encuentra el valor mínimo y máximo de la serie y sus índices correspondientes utilizando los métodos `min()`, `max()`, `idxmin()`, e `idxmax()`.


````python
print(f"Minimo = {min(pandasDic.values)}")
print(f"Maximo = {max(pandasDic.values)}")
print(f"Indice maximo = {pandasDic.idxmax()}")
print(f"Indice minimo = {pandasDic.idxmin()}")
````

3. Calcula la media, la suma, y el número de elementos únicos en la serie.

````python
print(f"Media = {pandasDic.mean()}")
print(f"Suma = {pandasDic.sum()}")
print(f"Unicos = {pandasDic.unique()}")
````
