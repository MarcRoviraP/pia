13. Crear dos matrices compatibles para multiplicación. Implementar la multiplicación matricial sin utilizar la función np.dot() ni el operador @. Mostrar las matrices originales y el resultado de la multiplicación.

```python
def multiplicacionMatricial(m1,m2):
    if m1.shape[1] != m2.shape[0]:
        return "Las matrices no son compatibles para multiplicación"
    
    filas = m1.shape[0]
    columnas = m2.shape[1]
    comun = m1.shape[1]
    
    resultado = np.zeros((filas, columnas))
    
    for fila in range(filas):
        for columna in range(columnas):
            for c in range(comun):
                resultado[fila][columna] += m1[fila][c] * m2[c][columna]
                
    return f"El resultado de la multiplicación es {resultado}"
    
    

matriz = np.random.randint(10,size=[2,2])
matriz2 = np.random.randint(10,size=[2,2])

print(multiplicacionMatricial(matriz,matriz2))
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def image_to_numpy(image_path):
    """
    Abre una imagen desde la ruta especificada y la convierte en una matriz de NumPy.

    Args:
        image_path (str): Ruta al archivo de imagen.

    Returns:
        numpy.ndarray: Matriz de NumPy que representa la imagen.
    """
    # Abrir la imagen usando PIL
    image = Image.open(image_path)
    if image.mode != 'RGB':
        image = image.convert('RGB')

    # Convertir la imagen a un array de NumPy
    image_array = np.array(image)

    return image_array

def mostrar_imagen(image_array):
    """
    Muestra una imagen a partir de una matriz de NumPy.

    Args:
        image_array (numpy.ndarray): Matriz de NumPy que representa la imagen.
    """
    plt.imshow(image_array)
    plt.axis('off')  # Ocultar los ejes
    plt.show()

def numpy_to_image(image_array, save_path=None):
    """
    Convierte una matriz de NumPy en una imagen y la muestra en pantalla.
    Opcionalmente, guarda la imagen en la ruta especificada.

    Args:
        image_array (numpy.ndarray): Matriz de NumPy que representa la imagen.
        save_path (str, optional): Ruta donde se guardará la imagen. Por defecto es None.
    
    Returns:
        PIL.Image.Image: Objeto de imagen de PIL.
    """
    # Convertir la matriz de NumPy a una imagen usando PIL
    image = Image.fromarray(image_array.astype('uint8'))

    # Mostrar la imagen usando matplotlib
    mostrar_imagen(image_array)

    # Guardar la imagen si se proporciona una ruta
    if save_path:
        image.save(save_path)
        print(f"Imagen guardada en: {save_path}")

    return image

# Ejemplo de uso
ruta_imagen = 'seta.png'  # Reemplaza con la ruta a tu imagen
imagen_numpy = image_to_numpy(ruta_imagen)


# Mostrar información de la matriz
'''print("Forma de la matriz:", imagen_numpy.shape)
print("Tipo de datos:", imagen_numpy.dtype)
print("Matriz de la imagen:\n", imagen_numpy)

# Mostrar la imagen
mostrar_imagen(imagen_numpy)

# Convertir la matriz de nuevo a una imagen y mostrarla
print("\nMostrando la imagen invertida utilizando numpy_to_image:")
imagen_invertida_pil = numpy_to_image(imagen_numpy)

# Opcional: Guardar la imagen
numpy_to_image(imagen_numpy, save_path='seta2.png')'''


#Girar imagen 90º Derecha
numpy_to_image(np.rot90(imagen_numpy,k=-1))

mostrar_imagen(imagen_numpy)
#Zona ojos
zonaOjos = imagen_numpy[336:400, 190:305] # seleccionar subzona
mostrar_imagen(zonaOjos)
# Crear máscara de píxeles negros (todos los canales = 0)
mask = np.all(zonaOjos == [0, 0, 0], axis=-1)  # True donde el píxel es negro
zonaOjos[mask] = [255,255,0] #Ojos amarillos

imagen_numpy[336:400, 190:305] = zonaOjos
mostrar_imagen(imagen_numpy)
rojo = zonaOjos.copy()
rojo[mask] = [255,0,0]
azul = zonaOjos.copy()
azul[mask] = [0,0,255]
verde = zonaOjos.copy()
verde[mask] = [0,255,0]
amarillo = zonaOjos.copy()
amarillo[mask] = [255,255,0] 

ojosRojos = imagen_numpy.copy()
ojosRojos[336:400, 190:305] = rojo
ojosAzules = imagen_numpy.copy()
ojosAzules[336:400, 190:305] = azul
ojosVerdes = imagen_numpy.copy()
ojosVerdes[336:400, 190:305] = verde
ojosAmarillos = imagen_numpy.copy()
ojosAmarillos[336:400, 190:305] = amarillo
fila1 = np.concatenate([ojosRojos,ojosAzules],axis=1)
fila2 = np.concatenate([ojosVerdes,ojosAmarillos],axis=1)
mostrar_imagen(np.concatenate([fila1,fila2]))
```
