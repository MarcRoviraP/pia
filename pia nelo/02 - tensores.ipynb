9. Crea un tensor 3D de forma (3, 4, 5) con valores aleatorios y reemplaza todos los elementos que están en posiciones donde la suma de los índices es un número par por el valor -1.


```python
tensor = torch.randn([3,4,5])
dim,j,k = torch.meshgrid(
    torch.arange(3),
    torch.arange(4),
    torch.arange(5)
)
mascara = (dim + j + k) % 2 == 0
tensor[mascara] = 0 
print(tensor)
```

```python
def dimPurge(t):
    t = t.squeeze()
    t = t.unsqueeze(0)
    t = t.unsqueeze(-1)
    return t
    
t = torch.randn([1,5,9,8,4,6,2,5,1,6,4,5])
print(dimPurge(t).shape)
```

8. Dado un tensor 3D de forma (3, 6, 9), escribir una función que lo divida en tensores más pequeños de forma (3, 2, 3) a lo largo de las dimensiones 1 y 2, y luego procese cada uno de los pequeños tensores individualmente.


```python
def procesarTensores(tensor):
    
    split = torch.split(tensor, tensor.size()[1]//3,dim=1)
    split2 = [torch.split(tensor, tensor.size()[2]//3,dim=2) for tensor in split] 
    for tupla in split2:
        for i in tupla:
            print(i.shape)
    
tensor = torch.randint(10,size=[3,6,9])
procesarTensores(tensor)
```

# Tensores imágenes

Dado el siguiente código que permite transformar imágenes en tensores y viceversa, realiza las operaciones que hemos hecho con la seta en numpy pero con tensores, sin pasar estos tensores a numpy.

```python
import torch
from PIL import Image
import torchvision.transforms as transforms
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def image_to_tensor_pytorch(image_path):
    """
    Carga una imagen desde la ruta especificada y la convierte en un tensor de PyTorch.

    Args:
        image_path (str): Ruta al archivo de imagen.

    Returns:
        torch.Tensor: Tensor que representa la imagen con forma (channels, height, width).
    """
    # Cargar la imagen utilizando PIL
    image = Image.open(image_path).convert('RGB')

    # Definir una transformación que convierte la imagen a tensor y normaliza los valores a [0, 1]
    transform = transforms.ToTensor()

    # Aplicar la transformación
    tensor = transform(image)

    return tensor

def tensor_pytorch_to_image(tensor, save_path=None):
    """
    Convierte un tensor de PyTorch de vuelta a una imagen y la muestra en pantalla.
    Opcionalmente, guarda la imagen en la ruta especificada.

    Args:
        tensor (torch.Tensor): Tensor que representa la imagen con forma (channels, height, width).
        save_path (str, optional): Ruta donde se guardará la imagen. Por defecto es None.

    Returns:
        PIL.Image.Image: Objeto de imagen de PIL.
    """
    # Asegurarse de que el tensor está en CPU y convertirlo a NumPy
    tensor = tensor.cpu()
    image_array = tensor.permute(1, 2, 0).numpy()  # Cambiar a (height, width, channels)

    # Convertir los valores a [0, 255] y tipo uint8
    image_array = np.clip(image_array * 255, 0, 255).astype(np.uint8)

    # Crear una imagen PIL desde el array de NumPy
    image = Image.fromarray(image_array)

    # Mostrar la imagen usando Matplotlib
    plt.imshow(image)
    plt.axis('off')  # Ocultar los ejes
    plt.show()

    # Guardar la imagen si se proporciona una ruta
    if save_path:
        image.save(save_path)
        print(f"Imagen guardada en: {save_path}")

    return image

def image_to_tensor_tensorflow(image_path):
    """
    Carga una imagen desde la ruta especificada y la convierte en un tensor de TensorFlow.

    Args:
        image_path (str): Ruta al archivo de imagen.

    Returns:
        tf.Tensor: Tensor que representa la imagen con forma (height, width, channels).
    """
    # Leer el archivo de imagen
    image_string = tf.io.read_file(image_path)

    # Decodificar la imagen (ajustar el formato según sea JPEG o PNG)
    image = tf.image.decode_image(image_string, channels=3)

    # Convertir la imagen a float32 y escalar los valores de píxeles a [0, 1]
    image = tf.image.convert_image_dtype(image, tf.float32)

    return image


def tensor_tensorflow_to_image(tensor, save_path=None):
    """
    Convierte un tensor de TensorFlow de vuelta a una imagen y la muestra en pantalla.
    Opcionalmente, guarda la imagen en la ruta especificada.

    Args:
        tensor (tf.Tensor): Tensor que representa la imagen con forma (height, width, channels).
        save_path (str, optional): Ruta donde se guardará la imagen. Por defecto es None.

    Returns:
        PIL.Image.Image: Objeto de imagen de PIL.
    """
    # Convertir el tensor a NumPy
    image_array = tensor.numpy()

    # Escalar los valores de [0, 1] a [0, 255] y convertir a uint8
    image_array = np.clip(image_array * 255, 0, 255).astype(np.uint8)

    # Crear una imagen PIL desde el array de NumPy
    image = Image.fromarray(image_array)

    # Mostrar la imagen usando Matplotlib
    plt.imshow(image)
    plt.axis('off')  # Ocultar los ejes
    plt.show()

    # Guardar la imagen si se proporciona una ruta
    if save_path:
        image.save(save_path)
        print(f"Imagen guardada en: {save_path}")

    return image

# Ruta a la imagen
ruta_imagen = 'ruta/a/tu/imagen.jpg'

# Convertir la imagen a tensor pytorch
tensor_imagen = image_to_tensor_pytorch(ruta_imagen)

# Convertir la imagen a tensor tensorflow
tensor_imagen = image_to_tensor_tensorflow(ruta_imagen)


print("Forma del tensor:", tensor_imagen.shape)
print("Tipo de datos:", tensor_imagen.dtype)


# Mostrar la imagen desde el tensor de PyTorch
print("\nMostrando la imagen original desde PyTorch tensor:")
tensor_pytorch_to_image(tensor_imagen)

print("\nMostrando la imagen desde TensorFlow tensor:")
tensor_tensorflow_to_image(tensor_imagen)

# Opcional: Guardar las imágenes 
tensor_pytorch_to_image(tensor_imagen, save_path='seta_pytorch.png')
tensor_tensorflow_to_image(tensor_imagen, save_path='seta_tensorflow.png')
```
